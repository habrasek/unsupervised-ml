```python
import pandas as pd
```

```python
df = pd.read_csv("crypto_data.csv")

df
```

```python
df = df.loc[df["IsTrading"] == True, :]

df.info()
```

```python
df = df.dropna()

df.info()
```

```python
df = df.loc[df["TotalCoinsMined"] > 0, :]

df.info()
```

```python
del df['CoinName']
```

```python
df
```

```python
dum_df = pd.get_dummies(df, columns=['Algorithm', 'ProofType'],prefix='Algorithm',drop_first=True)

del dum_df['Unnamed: 0']

del dum_df['IsTrading']

dum_df
```

```python
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dum_scaler = StandardScaler().fit_transform(dum_df[['TotalCoinsMined','TotalCoinSupply']])
```

```python
pca = PCA(n_components=0.99).fit_transform(dum_scaler)
```

```python
df_pca = pd.DataFrame(
    data=pca )
df_pca
```

```python
pca.explained_variance_ratio_
```

```python
inertia = []
k = list(range(1, 10))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df_pca)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(10)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
model = KMeans(n_clusters=5, random_state=5)

#fit the model
model.fit(df_pca)
#predict clusters
predictions = model.predict(df_pca)
len(predictions)
```

```python
df['prediction'] = predictions
```

```python
df.head(15)
```

```python
df['prediction'].value_counts()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(learning_rate=35)

tsne_features = tsne.fit_transform(df_pca)
```

```python
xs = []
ys = []
for t in tsne_features:
    xs.append(t[0])
    ys.append(t[1])
```

```python
plt.scatter(xs, ys)
plt.show()
```

```python
plt.scatter(xs, ys, c = predictions)
plt.show()
```
